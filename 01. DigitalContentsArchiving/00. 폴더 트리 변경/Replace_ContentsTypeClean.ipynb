```python
import os
import shutil

# 기본 경로 설정
contenttype_clean_path = 'D:/01. 빌리프_컨텐츠아카이빙/ContentType_Clean'  
brand_root_path = 'D:/01. 빌리프_컨텐츠아카이빙/Belif'  

# 옮기고 싶은 폴더 목록
folders_to_move = ['BasicAsset', 'SourceFile', 'EditedContent']

for product_line in os.listdir(brand_root_path):
        product_line_path = os.path.join(brand_root_path, product_line)
        
        # "0_BrandAsset" 폴더는 건너뛰기
        if product_line == "0_BrandAsset":
            print(f"Skipping brand asset folder: {product_line}")
            continue
        
        # ProductLine 폴더 내 ProductName 폴더 찾기
        if os.path.isdir(product_line_path):
            for product_name in os.listdir(product_line_path):
                product_name_path = os.path.join(product_line_path, product_name)
                if not os.path.isdir(product_name_path) or product_name == '0_Logo' :
                     continue
                
                for folder in folders_to_move :
                    src = os.path.join(contenttype_clean_path, folder)
                    dest = os.path.join(product_name_path, folder)
                     
                    if os.path.exists(src):
                        # 폴더 복사 (기존에 존재하면 덮어쓰기)
                        if os.path.exists(dest):
                              shutil.rmtree(dest)
                    shutil.copytree(src, dest)
                    print(f"Moved {folder} to {dest}")
```
