### 250408 변경 사항

```python
import os
import shutil

# 1) ProductType 위계 폴더 삭제
def remove_OldProductType(base_directory):
    for product_line in os.listdir(base_directory):
        product_line_path = os.path.join(base_directory, product_line)

        # "0_BrandAsset_브랜드자산" 폴더는 건너뛰기
        if product_line == "0_BrandAsset_브랜드자산":
            print(f"Skipping brand asset folder: {product_line}")
            continue

        elif product_line == "1_EditionSet_기획세트":
            print(f"Skipping EditionSet folder: {product_line}")
            continue

        # BrandLine 폴더 내 BrandName 폴더 찾기
        elif os.path.isdir(product_line_path):
            for product_name in os.listdir(product_line_path):
                product_name_path = os.path.join(product_line_path, product_name)
                print(f"Processing: {product_name}")
                for product_type_folder in os.listdir(product_name_path) :
                    product_type_folder_path = os.path.join(product_name_path, product_type_folder)
                    if os.path.isdir(product_type_folder_path) :
                        shutil.rmtree(product_type_folder_path)
                        print(f"삭제됨: {product_name} - {product_type_folder}")

def create_NewProductType(base_directory, source_dir):
    for product_line in os.listdir(base_directory):
        product_line_path = os.path.join(base_directory, product_line)

        # "0_BrandAsset_브랜드자산" 폴더는 건너뛰기
        if product_line == "0_BrandAsset_브랜드자산":
            print(f"Skipping brand asset folder: {product_line}")
            continue

        elif product_line == "1_EditionSet_기획세트":
            print(f"Skipping EditionSet folder: {product_line}")
            continue

        # BrandLine 폴더 내 BrandName 폴더 찾기
        elif os.path.isdir(product_line_path):
            for product_name in os.listdir(product_line_path):
                product_name_path = os.path.join(product_line_path, product_name)
                print(f"Processing: {product_name}")
                if product_name == '0_Logo' :
                    print(f"Skipping Logo folder : {product_name}")
                else :
                    for product_type_folder in os.listdir(source_dir) :
                        src = os.path.join(source_dir, product_type_folder)
                        dst = os.path.join(product_name_path, product_type_folder)
                        if os.path.isdir(src) :
                            shutil.copytree(src, dst, dirs_exist_ok= True) # dir이 이미 있어서 dirs_exist_ok 옵션이 없으면 오류 발생(내가 기존 폴더를 삭제안한 경우)
                            print(f"복사-이동됨: {product_type_folder} -> {product_name}")
                
```

### 250409 변경 사항

- 폴더 구조에 맞게 재이사

```python
import os
import shutil

content_mapping = {
    'BasicAsset': '1_BasicAsset_기본자산',
    'SourceFile': '2_SourceFile_원본자료',
    'EditedContent': '3_EditedContent_가공컨텐츠'
}

subfolder_mapping = {
    'Image': 'Image_이미지',
    'Video': 'Video_영상'
}

special_folders = ['Duo_듀오', 'Jumbo_점보', 'SingleEdition_단품기획']
special_destination = '4_SpecialSet_스페셜세트'

def copy_contents(src_dir, dst_dir):
    for content_type, dst_folder in content_mapping.items():
        src_content_path = os.path.join(src_dir, content_type)
        dst_content_path = os.path.join(dst_dir, dst_folder)

        if not os.path.exists(src_content_path):
            continue
        os.makedirs(dst_content_path, exist_ok=True)

        for item in os.listdir(src_content_path):
            src_item_path = os.path.join(src_content_path, item)

            # SourceFile 내부 폴더 이름 매핑 (Image -> Image_이미지 등)
            if content_type == 'SourceFile' and item in subfolder_mapping:
                mapped_folder = subfolder_mapping[item]
                mapped_dst_path = os.path.join(dst_content_path, mapped_folder)
                os.makedirs(mapped_dst_path, exist_ok=True)

                for subitem in os.listdir(src_item_path):
                    sub_src = os.path.join(src_item_path, subitem)
                    sub_dst = os.path.join(mapped_dst_path, subitem)

                    if os.path.exists(sub_dst):
                        if os.path.isdir(sub_dst):
                            shutil.rmtree(sub_dst)
                        else:
                            os.remove(sub_dst)

                    if os.path.isdir(sub_src):
                        shutil.copytree(sub_src, sub_dst)
                    else:
                        shutil.copy2(sub_src, sub_dst)
                continue

            dst_item_path = os.path.join(dst_content_path, item)

            if os.path.exists(dst_item_path) and os.path.isdir(dst_item_path):
                for subitem in os.listdir(src_item_path):
                    sub_src = os.path.join(src_item_path, subitem)
                    sub_dst = os.path.join(dst_item_path, subitem)

                    if os.path.exists(sub_dst):
                        if os.path.isdir(sub_dst):
                            shutil.rmtree(sub_dst)
                        else:
                            os.remove(sub_dst)

                    if os.path.isdir(sub_src):
                        shutil.copytree(sub_src, sub_dst)
                    else:
                        shutil.copy2(sub_src, sub_dst)
            else:
                if os.path.isdir(src_item_path):
                    shutil.copytree(src_item_path, dst_item_path)
                else:
                    shutil.copy2(src_item_path, dst_content_path)

    # 스페셜세트 복사 처리
    for special in special_folders:
        special_src = os.path.join(src_dir, special)
        if os.path.exists(special_src):
            special_dst = os.path.join(dst_dir, special_destination)
            os.makedirs(special_dst, exist_ok=True)

            target_path = os.path.join(special_dst, special)
            if os.path.exists(target_path):
                shutil.rmtree(target_path)

            shutil.copytree(special_src, target_path)
```

```python
# BasicAsset, EditedContent가 이동이 잘 안된 것 같음 + SpecialSet도

# 상위 경로 정의
src_base_directory = 'E:/01. 빌리프_컨텐츠아카이빙/Belif'
dst_base_directory = 'E:/01. 빌리프_컨텐츠아카이빙/Belif_Clean - 복사본'

# 순회하면서 src/dst 경로 설정
for src_product_line in os.listdir(src_base_directory):
    src_product_line_path = os.path.join(src_base_directory, src_product_line)

    if src_product_line in ["0_BrandAsset_브랜드자산", "1_EditionSet_기획세트"]:
        print(f"Skipping folder: {src_product_line}")
        continue

    if os.path.isdir(src_product_line_path):
        for src_product_name in os.listdir(src_product_line_path):
            src_product_name_path = os.path.join(src_product_line_path, src_product_name)

            # 실제 source, destination 경로 구성
            src_dir = product_name_path
            dst_dir = os.path.join(dst_base_directory, src_product_line, src_product_name)
            os.makedirs(dst_dir, exist_ok=True)

            print(f"Copying from: {src_product_name} → {dst_dir}")
            copy_contents(src_product_name_path, dst_dir)
```
