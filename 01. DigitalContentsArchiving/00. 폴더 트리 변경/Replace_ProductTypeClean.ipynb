### 250401 변경 사항

- 듀오, 점보, 단품, 단품 기획 폴더 추가

```python
import os
import shutil

# 기본 경로 설정
producttype_clean_path = 'E:/01. 빌리프_컨텐츠아카이빙/ProductType_Clean'  
brand_root_path = 'E:/01. 빌리프_컨텐츠아카이빙/Belif_Clean'  

# 옮기고 싶은 폴더 목록
folders_to_move = ['Duo_듀오', 'Jumbo_점보', 'Single_단품', 'SingleEdition_단품기획']

for product_line in os.listdir(brand_root_path):
        product_line_path = os.path.join(brand_root_path, product_line)
        
        # "0_BrandAsset" 폴더는 건너뛰기
        if product_line == "0_BrandAsset":
            print(f"Skipping brand asset folder: {product_line}")
            continue
        
        if product_line == '1_EditionSet_기획세트' : 
             print(f'Skipping EditionsSet folder: {product_line}')
             continue
      
        # ProductLine 폴더 내 ProductName 폴더 찾기
        # 1) 기존 ContentType Depth 삭제
        # 2) ProductType 대체
        if os.path.isdir(product_line_path):
            for product_name in os.listdir(product_line_path):
                product_name_path = os.path.join(product_line_path, product_name)
                if not os.path.isdir(product_name_path) or product_name == '0_Logo' :
                     continue

                shutil.rmtree(os.path.join(product_name_path, 'BasicAsset'))
                shutil.rmtree(os.path.join(product_name_path, 'SourceFile'))
                shutil.rmtree(os.path.join(product_name_path, 'EditedContent')) 

                for folder in folders_to_move :
                    src = os.path.join(producttype_clean_path, folder)
                    dest = os.path.join(product_name_path, folder)

                    if os.path.exists(src):
                        # 폴더 복사 (기존에 존재하면 덮어쓰기)
                        if os.path.exists(dest):
                              shutil.rmtree(dest)
                    shutil.copytree(src, dest)
                    print(f"Moved {folder} to {dest}")
```

```python
import os, shutil
```

```python
def create_folders(base_folder, folders):
    for folder in folders:
        path = os.path.join(base_folder, folder)
        os.makedirs(path, exist_ok=True)

def process_all_product_folders(base_directory):
    for brand_line in os.listdir(base_directory):
        brand_line_path = os.path.join(base_directory, brand_line)

        # "0_BrandAsset" 폴더는 건너뛰기
        if brand_line == "0_BrandAsset":
            print(f"Skipping brand asset folder: {brand_line}")
            continue
        
        # BrandLine 폴더 내 BrandName 폴더 찾기
        if os.path.isdir(brand_line_path):
            for product_name in os.listdir(brand_line_path):
                if product_name == '0_Logo' : 
                    print(f"Skipping ProductName - 0_Logo : {product_name}")
                    continue  
                product_name_path = os.path.join(brand_line_path, product_name)
                print(f"Processing: {product_name}")
                for content_type in os.listdir(product_name_path) :
                    content_type_path = os.path.join(product_name_path, content_type)
                    for edited_type in os.listdir(content_type_path) :
                        if edited_type == 'Store' :                    
                            store_path = os.path.join(content_type_path, 'Store')
                            create_folders(store_path, ['Wide', 'VMD'])
```

```python
def process_all_product_folders(base_directory):
    for brand_line in os.listdir(base_directory):
        brand_line_path = os.path.join(base_directory, brand_line)

        # "0_BrandAsset_브랜드자산" 폴더는 건너뛰기
        if brand_line == "0_BrandAsset_브랜드자산":
            print(f"Skipping brand asset folder: {brand_line}")
            continue
        
        # "1_EditionSet_기획세트" 폴더는 건너뛰기
        if brand_line == '1_EditionSet_기획세트' :
            for year in os.listdir(brand_line_path):
                year_path = os.path.join(brand_line_path, year)
                print(f"Processing: {year_path}")
                for plan_type in os.listdir(year_path) :
                    plan_type_path = os.path.join(year_path, plan_type)
                    for edition_type in os.listdir(plan_type_path) :
                        edition_type_path = os.path.join(plan_type_path, edition_type)
                        for content_type in os.listdir(edition_type_path) :
                            if content_type == 'EditedContent' :
                                content_type_path = os.path.join(edition_type_path, content_type)
                                for edited_type in os.listdir(content_type_path) :
                                    if edited_type == 'Store' :                  
                                        store_path = os.path.join(content_type_path, 'Store')
                                        create_folders(store_path, ['Wide', 'VMD'])
                
        # BrandLine 폴더 내 BrandName 폴더 찾기
        if os.path.isdir(brand_line_path):
            for product_name in os.listdir(brand_line_path):
                product_name_path = os.path.join(brand_line_path, product_name)
                print(f"Processing: {product_name}")
                for product_type in os.listdir(product_name_path) :
                    product_type_path = os.path.join(product_name_path, product_type)
                    for content_type in os.listdir(product_type_path) :
                        content_type_path = os.path.join(product_type_path, content_type)
                        for edited_type in os.listdir(content_type_path) :
                            if edited_type == 'Store' :                    
                                store_path = os.path.join(content_type_path, 'Store')
                                create_folders(store_path, ['Wide', 'VMD'])
```

### 250408 변경 사항

- 폴더 구조 1) BasicAsset 2) SourceFiles 3) EditedContents 4) SpecialSet

```python
import os
import shutil

# 1) ProductType 위계 폴더 삭제
def remove_OldProductType(base_directory):
    for product_line in os.listdir(base_directory):
        product_line_path = os.path.join(base_directory, product_line)

        # "0_BrandAsset_브랜드자산" 폴더는 건너뛰기
        if product_line == "0_BrandAsset_브랜드자산":
            print(f"Skipping brand asset folder: {product_line}")
            continue

        elif product_line == "1_EditionSet_기획세트":
            print(f"Skipping EditionSet folder: {product_line}")
            continue

        # BrandLine 폴더 내 BrandName 폴더 찾기
        elif os.path.isdir(product_line_path):
            for product_name in os.listdir(product_line_path):
                product_name_path = os.path.join(product_line_path, product_name)
                print(f"Processing: {product_name}")
                for product_type_folder in os.listdir(product_name_path) :
                    product_type_folder_path = os.path.join(product_name_path, product_type_folder)
                    if os.path.isdir(product_type_folder_path) :
                        shutil.rmtree(product_type_folder_path)
                        print(f"삭제됨: {product_name} - {product_type_folder}")

def create_NewProductType(base_directory, source_dir):
    for product_line in os.listdir(base_directory):
        product_line_path = os.path.join(base_directory, product_line)

        # "0_BrandAsset_브랜드자산" 폴더는 건너뛰기
        if product_line == "0_BrandAsset_브랜드자산":
            print(f"Skipping brand asset folder: {product_line}")
            continue

        elif product_line == "1_EditionSet_기획세트":
            print(f"Skipping EditionSet folder: {product_line}")
            continue

        # BrandLine 폴더 내 BrandName 폴더 찾기
        elif os.path.isdir(product_line_path):
            for product_name in os.listdir(product_line_path):
                product_name_path = os.path.join(product_line_path, product_name)
                print(f"Processing: {product_name}")
                if product_name == '0_Logo' :
                    print(f"Skipping Logo folder : {product_name}")
                    continue
                for product_type_folder in os.listdir(product_name_path) :
                    product_type_folder_path = os.path.join(product_name_path, product_type_folder)
                        src = os.path.join(source_dir, product_type_folder)
                        dst = product_type_folder_path
                        if os.path.isdir(src) :
                            shutil.copytree(src, dst, dirs_exist_ok= True)
                            print(f"복사-이동됨: {product_type_folder} -> {product_name}")
                
```

```python
# 경로 설정
base_directory = "E:/01. 빌리프_컨텐츠아카이빙/Belif_Clean - 복사본"
source_dir = "E:/01. 빌리프_컨텐츠아카이빙/ProductType_Clean"  # 옮기고 싶은 폴더들이 있는 경로
```

```python
remove_OldProductType(base_directory)
```

```python
create_NewProductType(base_directory, source_dir)
```
