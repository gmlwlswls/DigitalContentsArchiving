```python
# 확장자별 메타데이터 입력 여부 다름

# 1. .jpg : 탐색기 > 속성 > 자세히 탭에서 설명 필드에 표시
# 2. mp4, mov : 탐색기 > 속성 > 자세히 탭에서 제목 필드에 표시
# 2. .psd, .ai : Adobe Bridge와 같은 전용 툴에서만 자세히 보임, Adobe Bridge Description 필드에서 확인 가능
# 3. .png, .gif, .ssg : 탐색기에서 확인 불가(백업 파일이나 엑셀 매핑 권장)

import os
import re
import subprocess

# 📂 메타데이터 입력 대상 확장자 목록
valid_exts = [".jpg", ".psd", ".ai", ".mp4", ".mov"]

# 📁 순회할 루트 디렉토리 설정 (예: 'D:/MyFolder/Belif')
target_dir = r"D:/00. Test/Belif"  # ← 여기에 본인 폴더 경로 입력

# 📄 문서번호 패턴 (예: 'DOC00045_...')
docnum_pattern = re.compile(r'^(DOC\d{4,5})_')

# 📌 exiftool 실행 경로 (환경변수 등록 안 돼 있으면 절대경로 지정)
exiftool_cmd = "C:/Tools/ExifTool/exiftool"

# 🔁 폴더 내 전체 파일 순회
for root, _, files in os.walk(target_dir):
    for file in files:
        file_path = os.path.join(root, file)
        name, ext = os.path.splitext(file)
        ext = ext.lower()

        if ext in valid_exts:
            # 문서번호 추출
            match = docnum_pattern.match(file)
            if match:
                docnum = match.group(1)
                print(f"처리 중: {file} → 문서번호: {docnum}")

                # 확장자별 필드 설정
                if ext == ".jpg":
                    args = [f"-ImageDescription={docnum}"]
                elif ext in [".ai", ".psd"]:
                    args = [f"-XMP:Description={docnum}"]
                elif ext in [".mp4", ".mov"]:
                    args = [f"-Title={docnum}"]
                else:
                    continue  # 안전망

                # exiftool 실행
                subprocess.run([
                    exiftool_cmd,
                    *args,
                    "-overwrite_original",
                    file_path
                ])
            else:
                print(f"[스킵] 문서번호 없음: {file}")
```
